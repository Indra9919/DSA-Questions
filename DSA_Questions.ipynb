1. Check if a Number is Prime: 
Problem: Write a function to check if a given number is prime. 
Input: Number: 29 
Output: True 
Explanation: 29 is a prime number.

```python
def is_prime(n):
    if n <= 1:
        return "Not Prime"

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "Not Prime"

    return "Prime"


num = int(input("Enter number: "))
print(is_prime(num))
```
